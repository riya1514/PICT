```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
# the distribution of the sepal length in the iris, with 10 bins to display the frequency of different sepal length values.
sns.histplot(data=iris, x='SepalLengthCm',bins=10)
plt.show()
```

```python
# the distribution of the sepal width in the iris, with 10 bins to display the frequency of different sepal length values.
sns.histplot(data=iris, x='SepalWidthCm', bins=10)
plt.show()
```

```python
# the distribution of the petal length in the iris, with 10 bins to display the frequency of different sepal length values.
sns.histplot(data=iris, x='PetalLengthCm', bins=15)
plt.show()
```

```python
# the distribution of the petal width in the iris, with 10 bins to display the frequency of different sepal length values.
sns.histplot(data=iris, x='PetalWidthCm', bins=20)
plt.show()
```

```python
# distribution of sepal length for each species in the dataset, highlighting the median, quartiles, 
# and any potential outliers for sepal length within each species.
sns.boxplot(data=iris, x="Species", y="SepalLengthCm")
plt.show()
```

```python
# distribution of sepal width for each species in the dataset, highlighting the median, quartiles, 
# and any potential outliers for sepal length within each species.
sns.boxplot(data=iris,x="Species", y="SepalWidthCm")
plt.show()
```

```python
 # distribution of Petal length for each species in the dataset, highlighting the median, quartiles, 
# and any potential outliers for sepal length within each species.
sns.boxplot(data=iris, x="Species", y="PetalLengthCm")
plt.show()
```

```python
# distribution of petal width for each species in the dataset, highlighting the median, quartiles, 
# and any potential outliers for sepal length within each species.
sns.boxplot(data=iris,x="Species", y="PetalWidthCm")
plt.show()
```

```python
# estimated probability density of the sepal length in the dataset, providing a smooth curve to visualize the distribution of sepal length values.
sns.kdeplot(data=iris, x="SepalLengthCm")
plt.show()
```

```python
# estimated probability density of the sepal width in the dataset, providing a smooth curve to visualize the distribution of sepal length values.
sns.kdeplot(data=iris, x="SepalWidthCm")
plt.show()
```

```python
# estimated probability density of the petal length in the dataset, providing a smooth curve to visualize the distribution of sepal length values.
sns.kdeplot(data=iris, x="PetalLengthCm")
plt.show()
```

```python
# estimated probability density of the petal width in the dataset, providing a smooth curve to visualize the distribution of sepal length values.
sns.kdeplot(data=iris, x="PetalWidthCm")
plt.show()
```

```python
# Create a box plot for each feature
sns.boxplot(data=iris.drop(["Species"],axis=1))
plt.ylim(0, 10)
plt.title('Boxplot of Features in the Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris.drop(columns='Species'))
plt.title('Violin Plot of Features in the Iris Dataset')
plt.show()
```

```python
# Scatter plot with hue by species (to see the impact of outliers in the context of species)
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot with Outliers Visualization')
plt.show()
```

```python
corr_matrix = iris[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].corr()
```

```python
# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.show()
```
