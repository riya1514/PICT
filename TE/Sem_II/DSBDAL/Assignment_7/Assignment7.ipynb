```python
import nltk
```

```python
nltk.download('popular')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
with open("data.txt","r") as file:
    p1 = file.read()

with open("data2.txt","r") as file:
    p2 = file.read()

# Alternate method
# p1 = '''The Department of Computer Engineering was established in 1983 with the intent to provide quality education to the young aspirants.'''
```

```python
data = [p1]
```

```python
from nltk import sent_tokenize

sentence_tokens = sent_tokenize(p1)
for sentence in sentence_tokens:
    print(sentence)
sentence_tokens
```

```python
from nltk import word_tokenize
word_tokens = word_tokenize(p1)
for word in word_tokens:
    print(word)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')

filtered_word_tokens = []

for tokens in word_tokens:
        if tokens.lower() not in stopwords:  # optional: .lower() for case insensitivity
            filtered_word_tokens.append(tokens)

for filtered in filtered_word_tokens:
    print(filtered)
```

```python
from nltk import pos_tag
pos_tags = pos_tag(filtered_word_tokens)
print(pos_tags)
```

```python
#Stemming 
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_words=[]
for word in filtered_word_tokens:
    stemmed_words.append(ps.stem(word))
print(stemmed_words)
```

```python
#Lemmatization 
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words=[]
for word in filtered_word_tokens:
    lemmatized_words.append(lemmatizer.lemmatize(word))
print(lemmatized_words)
```

```python
from nltk import FreqDist
from sklearn.feature_extraction.text import TfidfVectorizer

# Example list of stemmed words
stemmed_words = p1.lower().split()  # This is an example, you would typically use stemming here

# 1. Calculate Term Frequency (TF) using FreqDist
tf = FreqDist(stemmed_words)

# 2. Prepare corpus for calculating IDF (a list of documents)
corpus = [p1,p2]  # In this example, we only have one document, but you can have more documents

# 3. Calculate IDF using TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(corpus)  # Fit the vectorizer to the corpus

# Retrieve the IDF values
idf = tfidf_vectorizer.idf_

# Map words to their corresponding IDF values
idf_dict = {word: idf[i] for i, word in enumerate(tfidf_vectorizer.get_feature_names_out())}

# 4. Calculate TF-IDF manually by multiplying TF and IDF
tfidf = {word: tf[word] * idf_dict.get(word, 0) for word in tf.keys()}  # Default 0 for words not in IDF

# Print results
print("Term Frequency (TF):", tf)
print("\n\nInverse Document Frequency (IDF):", idf_dict)
print("\n\nTF-IDF:", tfidf)

# Specify the word you want to find the TF-IDF for
specific_word = "computer"  # You can change this to any word you are interested in

# Fetch TF-IDF for the specific word
specific_word_tf = tf.get(specific_word, 0)  # Default to 0 if the word isn't found
print(f"\nTF for the word '{specific_word}':", specific_word_tf)

specific_word_idf = idf_dict.get(specific_word, 0)  # Default to 0 if the word isn't found
print(f"\nIDF for the word '{specific_word}':", specific_word_idf)

specific_word_tfidf = tfidf.get(specific_word, 0)  # Default to 0 if the word isn't found
print(f"\nTF-IDF for the word '{specific_word}':", specific_word_tfidf)
```

```python
#Analysis

# NLTK (Natural Language Toolkit).

# 1. Tokenization
# The text was successfully split into words and sentences.

# 2.POS (Parts of speech) tagging
# Words were correctly classified into nouns, verbs, adjectives, etc.
# DT = Determiner (e.g  The, A)
# JJ = Adjective (e.g  quick, lazy)
# NN = Noun (e.g  fox, dog)
# VBZ = Verb (e.g  jumps)
# IN = Preposition (e.g  over)

# 3.Stop word removal
# Common words like "is", "the", "and" were successfully removed.

# 4.Stemming
# Words were reduced to their root form (e.g  running -> run).
# Not really good for all words as it cuts word to its base form (e.g  files -> fli)

# 5.Lemmatization
# Better alternative foe stemming as it converts word to meaningful format
# Words were reduced to their dictionary form (e.g  flies -> fly)

# 6. TF : Measures how often a word appears in a document relative to the total number of words in that document.
# TF helps in understanding which words are more frequent or dominant within a specific document. However, it doesn't 
# account for the word's importance across multiple documents.

# 7. IDF :  Measures how important a word is by considering how rare or common it is across a corpus of documents. 
# Words that appear frequently across all documents are less important.
# IDF helps to adjust the weight of common words (like "the", "and", "is") by reducing their importance. It boosts the 
# value of rare words that are more likely to contain meaningful information.

# 8. Combines TF and IDF to give a score for each word, representing both its importance in a document (TF) and its rarity 
# across the entire corpus (IDF).
# TF-IDF is widely used in text mining and information retrieval to rank words by their relevance. Words that are frequent in 
# a document but rare across the corpus get high TF-IDF scores, making them highly significant for the document.
```
